```python
#image preprocessing
#1. Resizing
#2. Flattening
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread #for reading the image(opencv can be used as well)
from skimage.transform import resize

target = [] #empty output target list
images = [] #input
flat_data = [] #flattened vector

DATADIR = '/content/images'
CATEGORIES = ['Cat', 'Dog', 'Kitten', 'Puppy'] #target labels

for category in CATEGORIES:
  class_num = CATEGORIES.index(category) #Label encoding with the index values
  path = os.path.join(DATADIR, category) #create path to use all the images
  for img in os.listdir(path):
    img_array = imread(os.path.join(path, img))
    #print(img_array)
    img_resized = resize(img_array,(150, 150, 3)) #(Using skimage) Normalizes the pixel value from 0 to 1
    flat_data.append(img_resized.flatten())
    images.append(img_resized)
    target.append(class_num)

flat_data = np.array(flat_data)
target = np.array(target)
images = np.array(images)
```

```python
#len(flat_data[0]) is 150 * 150 * 3
target
```

```python
unique,count = np.unique(target, return_counts = True) #To Plot a graph
plt.bar(CATEGORIES,count)
```

```python
#Split data into training and testing (Train test split from model_selection of sklearn ml model)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(flat_data,target,test_size = 0.3, random_state = 109)
```

```python
#classification algorithm(Supervised learning method, labelled data) used - SVM (Support Vector Machine)
from sklearn.model_selection import GridSearchCV #Ensures that the parameters used in Support Vectors are Proper
from sklearn import svm
param_grid = [
              {'C':[1,10,100,1000], 'kernel':['linear']},
              {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']}
]

svc = svm.SVC(probability = True)
clf = GridSearchCV(svc, param_grid)
clf.fit(x_train, y_train)

```

```python
y_pred = clf.predict(x_test) #To check the accuracy of the model
y_pred #predicted output
```

```python
y_test #real output
```

```python
#Evaluate our model
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_pred, y_test)
```

```python
confusion_matrix(y_pred, y_test)
```

```python
# Save the model using Pickle library
import pickle
pickle.dump(clf, open('img_model.p','wb'))
```
